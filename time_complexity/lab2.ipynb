```python
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

input_sizes = [100, 500, 1000, 5000, 10000, 20000]
best_times, avg_times, worst_times = [], [], []

for n in input_sizes:
    arr = [random.randint(1, 100000) for _ in range(n)]
    
    # Best Case (first element)
    start = time.time()
    linear_search(arr, arr[0])
    best_times.append((time.time() - start) * 1000)  # ms
    
    # Average Case (middle element)
    start = time.time()
    linear_search(arr, arr[n//2])
    avg_times.append((time.time() - start) * 1000)
    
    # Worst Case (not present)
    start = time.time()
    linear_search(arr, -1)
    worst_times.append((time.time() - start) * 1000)


plt.plot(input_sizes, best_times, color="green", label='Best Case (O(1))',marker="o")
plt.plot(input_sizes, avg_times, color="blue", label='Avg Case (O(n))',marker="o")
plt.plot(input_sizes, worst_times, color='red', label='Worst Case (O(n))',marker="o")
plt.xlabel('Input Size (n)')
plt.ylabel('Time (milliseconds)')
plt.title('Linear Search Time Complexity Analysis')
plt.grid(True)
plt.show()

# Print results
print("Input Size (n) | Best Time (ms) | Avg Time (ms) | Worst Time (ms)")
print("--------------------------------------------------------------")
for i, n in enumerate(input_sizes):
    print(f"{n:14} | {best_times[i]:13.4f} | {avg_times[i]:12.4f} | {worst_times[i]:13.4f}")
print("\n--- Theoretical vs Empirical Analysis ---")
print("• Best Case    → 1 comparison (O(1))")
print("• Average Case → ~n/2 comparisons (O(n/2))")
print("• Worst Case   → n comparisons (O(n))")
print("• As n increases, comparisons grow linearly in average & worst case.")
```
