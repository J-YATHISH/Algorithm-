```python
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1


n = 10000
arr = [random.randint(1, 100000) for _ in range(n)]

#Best Case
key = arr[0]
start = time.time()
result = linear_search(arr, key)
best_case = (time.time() - start) * 1000 # in milliseconds

#Average Case
key = arr[n//2]
start = time.time()
result = linear_search(arr, key)
average_case = (time.time() - start) * 1000

# Worst case
key = -1  # Not in array
start = time.time()
result = linear_search(arr, key)
worst_case = (time.time() - start) * 1000

cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_case, average_case, worst_case]

plt.bar(cases, times)
plt.title('Linear Search Performance (n=10,000)')
plt.ylabel('Time (milliseconds)')
plt.show()

print(f"Best case time: {best_case:5f} ms")
print(f"Average case time: {average_case:.5f} ms")
print(f"Worst case time: {worst_case:.5f} ms")

print("\n--- Observations ---")
print(f"• With input size n = {n}, worst case time is highest.")
print("• Best case is nearly instantaneous (constant time).")
print("• Average case scales with list size, roughly half the work of the worst case.")
print("• As 'n' increases, worst and average cases grow linearly.")
```
